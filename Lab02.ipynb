### seqCleaner

```python
#!/usr/bin/env python3 
# Name: Akash Pandit (aspandit)
# Group Members: Preeti Rajarathinam(prrajara), Delson Hays (dhhays), Natalie Cellucci(ncellucc), Riya Tiloda (rtiloda)

'''
Read a DNA string from user input and return a collapsed substring of embedded Ns to: {count}.

Example: 
input: AaNNNNNNGTC
output: AA{6}GTC

Any lower case letters are converted to uppercase
'''

class DNAstring(str):  # gotta love inheritance where I can use string methods on this object B)
    """DNAstring class built off of the string class, DNAstring objects will have all the properties of strings + a length method which returns their length
    and a purify method which uppercases a string and condenses any blocks of N"""
    
    def length(self):
        '''return the length of the string'''
        return len(self)
    
    def purify(self):
        '''Return an upcased version of the string, collapsing a single run of Ns.
        Assumptions:
        - inputs consist solely of capital or lowercase A, G, C, T, or N
        - only 1 "block" of Ns appear per string'''
        self = self.upper()  # convert all string characters to uppercase 
        nCount = self.count('N')  # get how many Ns appear in the block of Ns, save that as nCount
        nIndex = self.index('N')  # find the index of the first occurrance of N in the string, save that as nIndex

        pureDNA = list(self.replace('N', ''))  # clear out the Ns and split up the string into a list (mutable), save that as pureDNA
        pureDNA.insert(nIndex, f'{{{nCount}}}')  # insert '{n_count}' at the index where the Ns began in pureDNA (list)

        return ''.join(pureDNA)  # return pureDNA joined together as a string


"""Driver"""
def main():
    ''' Get user DNA data and clean it up.'''
    data = input('DNA data?')
    while (data):
        thisDNA = DNAstring(data)  # get a raw string of dna
        pureData = thisDNA.purify()  # run the purify method on that raw string, storing the purified string as pureData
        print(pureData)
        data = input('DNA data?')
    
main()  # run the driver
```

### Hints: 
 - Each coordinate will contain only 3 numeric values. 
 - Bond lengths are the distance between two points, which for points P and Q in 3-space,
$(P_x , P_y , P_z )$ and $(Q_x , Q_y , Q_z )$ respectively, the distance between them is:


$$
\begin {eqnarray}
     \|PQ\| &=&\sqrt{((P_x-Q_x)^2 + (P_y-Q_y)^2 + (P_z-Q_z)^2)} \\
            &=&\sqrt{\sum^{i \in x,y,z}{(P_i-Q_i)^2}} \\
\end{eqnarray}
$$

 - Bond angles can be calculated from the dot product.
 
Let’s say that we have three points in space labeled P, Q and R. We are interested in the angle at point Q that is made from line segments QP and QR.
The dot product tells us (for standard vectors P and R) that:
$$P \cdot R=\|P\| \|R\| \cos{\theta}$$
in this notation,$\|P\|$ refers to the length of vector P as a standard vector (assumed to begin at the origin (0,0,0) ).
We can then see that the angle between vectors P and R can be found by:
$$\cos{\theta}=\frac{P \cdot R} {\|P\| \|R\|}$$

We can calculate the dot product using the sum of products of the vector components:
$$P \cdot R = \sum^{i \in x,y,z}{P_i R_i} $$

Now, to find vector P in standard form, we need to remember that QP starts at Q, so we need to place the origin at Q and find out where P is in that new space. We do that by subtracting the components of Q from P.
Putting all of this together, we get:
$$\theta=\cos^{-1}{\frac{\sum^{i \in x,y,z}{(P_i - Q_i) (R_i - Q_i) }} {\|QP\| \|QR\|}}$$
Remember, $\theta$ is in radians.

### coordinateMathSoln

```python
#!/usr/bin/env python3 
# Name: Akash Pandit (aspandit)
# Group Members: Preeti Rajarathinam(prrajara), Delson Hays (dhhays), Natalie Cellucci(ncellucc), Riya Tiloda (rtiloda)

'''
This program defines a class Triad with methods to get distance between 2 3-D points and the angle between 3 3-D points.
A driver function main is also written to parse user input to be used as parameters in a Triad object for manipulation by
its classes. 

A proper input would look like: 'C = (1x, 1y, 1z) N = (2x, 2y, 2z) Ca = (3x, 3y, 3z)'
with the Nxyz parameters representing coordinates for each atom C, N, or Ca.

A proper output, assuming N as the angle point, would display:
'N-C bond length = {float}'
'N-Ca bond length = {float}'
'C-N-Ca bond angle = {float}' 
'''

import math
class Triad :
    """
    Calculate angles and distances among a triad of points.
 
    Author: David Bernick
    Date: March 21, 2013
    Points can be supplied in any dimensional space as long as they are consistent.
    Points are supplied as tupels in n-dimensions, and there should be three
    of those to make the triad. Each point is positionally named as p,q,r
    and the corresponding angles are then angleP, angleQ and angleR.
    Distances are given by dPQ(), dPR() and dQR()
 
    Required Modules: math
    initialized: 3 positional tuples representing Points in n-space
             p1 = Triad( p=(1,0,0), q=(0,0,0), r=(0,1,0) )
    attributes: p,q,r the 3 tuples representing points in N-space
    methods:  angleP(), angleR(), angleQ() angles measured in radians
          dPQ(), dPR(), dQR() distances in the same units of p,q,r
 
    """
 
    def __init__(self,p,q,r) :
        """ Construct a Triad. 
        
        Example construction:
            p1 = Triad( p=(1.,0.,0.), q=(0.,0.,0.), r=(0.,0.,0.) ). 
        """
        self.p = p
        self.q = q
        self.r = r
        
# private helper methods
    def d2 (self,a,b) : # calculate squared distance of point a to b
        return float(sum((ia-ib)*(ia-ib)  for  ia,ib in zip (a,b)))
    
    def dot (self,a,b) : # dotProd of standard vectors a,b
        return float(sum(ia*ib for ia,ib in zip(a,b)))
    
    def ndot (self,a,b,c) : # dotProd of vec. a,c standardized to b
        return float(sum((ia-ib)*(ic-ib) for ia,ib,ic in zip (a,b,c)))
    
# calculate lengths(distances) of segments PQ, PR and QR
    def dPQ (self):
        """ Provides the distance between point p and point q """
        return math.sqrt(self.d2(self.p,self.q))
    
    def dPR (self):
        """ Provides the distance between point p and point r """
        return math.sqrt(self.d2(self.p,self.r))
    
    def dQR (self):
        """ Provides the distance between point q and point r """
        return math.sqrt(self.d2(self.q,self.r))
    
    def angleP (self) :
        """ Provides the angle made at point p by segments pq and pr (radians). """
        return math.acos(self.ndot(self.q,self.p,self.r) / 
                         math.sqrt(self.d2(self.q,self.p) * self.d2(self.r,self.p)))
    
    def angleQ (self) :
        """ Provides the angle made at point q by segments qp and qr (radians). """
        return math.acos(self.ndot(self.p,self.q,self.r) /
                         math.sqrt(self.d2(self.p,self.q) * self.d2(self.r,self.q)))
 
    def angleR (self) :
        """ Provides the angle made at point r by segments rp and rq (radians). """
        return math.acos(self.ndot(self.p,self.r,self.q) /
                         math.sqrt(self.d2(self.p,self.r) * self.d2(self.q,self.r)))

def main():
    '''Driver function- gather user input for coordinates, clean the data, then parse the distances and angles and output to stdout'''
    # get user input
    coords = input("Please input your coordinates!")

    # clean user input and turn it into a list of the floats
    charsToInclude = '0123456789. '
    coords = ''.join(char for char in coords if char in charsToInclude)  # rewrite string with only chars in charsToInclude
    coords = [float(num) for num in coords.strip().split()]  # split into list of the nums and convert them to floats

    # assign points to input to point vars & initialize Triad object with them
    point1, point2, point3 = tuple(coords[:3]), tuple(coords[3:6]), tuple(coords[6:9])
    triad = Triad(point1, point2, point3)

    radToDeg = 180 / math.pi  # conversion multiplier from radians to degrees

    print('N-C bond length =', f'{triad.dPQ():.2f}')  # prints bond length between N and C to 3 sig decimals
    print('N-Ca bond length =', f'{triad.dQR():.2f}')  # prints bond length between N and Ca to 3 sig decimals
    print('C-N-Ca bond angle =', round(triad.angleQ() * radToDeg, 0))  # prints the bond angle at Nitrogen


main()  # execute driver
```

```python
#!/usr/bin/env python3 
# Name: Akash Pandit
# Group Members: Preeti Rajarathinam(prrajara), Delson Hays (dhhays), Natalie Cellucci(ncellucc), Riya Tiloda (rtiloda)

'''
This program contains various dictionaries displaying the key:value relationships between amino acid representations
and codons, both DNA and RNA. A driver function main takes a user input and tests whether it is a key in any of the
given dictionaries and returns its appropriate value, or 'unknown' if nothing matches.

This is done through a series of conditional statements checking if an input is in a dictionary's keys, and returning
the appropriate value if so.

Some example inputs and their outputs in format:
input = output

C = CYS
GLY = G
UUU = PHE
ATC = Ile
0 = unknown
baller = unknown

The names of short_AA, long_AA, and RNA_codon_table changed to shortAA, longAA, and rnaCodonTable respectively
to match this class' request to use camelCase naming conventions.
'''
shortAA = {
            'CYS': 'C', 'ASP': 'D', 'SER': 'S', 'GLN': 'Q', 'LYS': 'K',
            'ILE': 'I', 'PRO': 'P', 'THR': 'T', 'PHE': 'F', 'ASN': 'N', 
            'GLY': 'G', 'HIS': 'H', 'LEU': 'L', 'ARG': 'R', 'TRP': 'W', 
            'ALA': 'A', 'VAL': 'V', 'GLU': 'E', 'TYR': 'Y', 'MET': 'M'
            }  # maps 3 char amino acid acronyms to 1 char acronyms

longAA = {value:key for key,value in shortAA.items()}  # maps 1 char amino acid acronyms to 3 char acronyms

rnaCodonTable = {  # maps RNA codons to their respective amino acids, with stop codons designated as ---
# Second Base
# U             C             A             G
#U
'UUU': 'Phe', 'UCU': 'Ser', 'UAU': 'Tyr', 'UGU': 'Cys',
'UUC': 'Phe', 'UCC': 'Ser', 'UAC': 'Tyr', 'UGC': 'Cys',
'UUA': 'Leu', 'UCA': 'Ser', 'UAA': '---', 'UGA': '---',
'UUG': 'Leu', 'UCG': 'Ser', 'UAG': '---', 'UGG': 'Trp',
#C 
'CUU': 'Leu', 'CCU': 'Pro', 'CAU': 'His', 'CGU': 'Arg',
'CUC': 'Leu', 'CCC': 'Pro', 'CAC': 'His', 'CGC': 'Arg',
'CUA': 'Leu', 'CCA': 'Pro', 'CAA': 'Gln', 'CGA': 'Arg',
'CUG': 'Leu', 'CCG': 'Pro', 'CAG': 'Gln', 'CGG': 'Arg',
#A
'AUU': 'Ile', 'ACU': 'Thr', 'AAU': 'Asn', 'AGU': 'Ser',
'AUC': 'Ile', 'ACC': 'Thr', 'AAC': 'Asn', 'AGC': 'Ser',
'AUA': 'Ile', 'ACA': 'Thr', 'AAA': 'Lys', 'AGA': 'Arg',
'AUG': 'Met', 'ACG': 'Thr', 'AAG': 'Lys', 'AGG': 'Arg',
#G
'GUU': 'Val', 'GCU': 'Ala', 'GAU': 'Asp', 'GGU': 'Gly',
'GUC': 'Val', 'GCC': 'Ala', 'GAC': 'Asp', 'GGC': 'Gly',
'GUA': 'Val', 'GCA': 'Ala', 'GAA': 'Glu', 'GGA': 'Gly',
'GUG': 'Val', 'GCG': 'Ala', 'GAG': 'Glu', 'GGG': 'Gly'
}
dnaCodonTable = {key.replace('U','T'):value for key, value in rnaCodonTable.items()}  
# same as rnaCodonTable but with T instead of U

def main() -> None:
    '''Driver function, prompts user for input then searches for that input in various dictionaries to output
    the input's codon or amino acid equivalent, printing unknown if nothing is found.'''
    inp = input('Search for your amino acid or codon here! ')  # prompt user to search for codon/AA

    if inp.upper in longAA.keys():
        print(inp.upper(), '=', longAA[inp.upper()].upper())  # the input was found in longAA

    elif inp.upper() in shortAA.keys():
        print(inp.upper(), '=', shortAA[inp.upper()].upper())  # the input was found in shortAA

    elif inp.upper() in rnaCodonTable.keys():
        print(inp.upper(), '=', rnaCodonTable[inp.upper()].upper())  # the input was found in rnaCodonTable

    elif inp.upper() in dnaCodonTable.keys():
        # for DNA inputs without T, they would be pulled from the RNA Codon Table as they would be identical to the DNA codon
        print(inp.upper(), '=', dnaCodonTable[inp.upper()].upper())  # the input has a T and was found in dnaCodonTable

    else:
        print('unknown')  # the input was not found in any of the dictionaries
    

main()
```
